```
using Microsoft.EntityFrameworkCore;
using KrakenTradesDatabase;

var db = new AppDbContext();
```

```
Trace candles_to_trace(string name, CandleUnit unit, int interval)
{
    var candles = db.Candles
        .Where(candle => candle.Symbol.Name == name)
        .Where(candle => candle.CandleUnit == unit)
        .Where(candle => candle.Interval == interval);
    
    return new Scattergl()
    {
        name = name,
        
        x = candles.Select(row => row.DateTimeOffset),
        y = candles.Select(row => row.Open)
    };
}
```

```
using XPlot.Plotly;
using XPlot.Plotly.Interactive;

var chart = Chart.Plot(new Trace [] 
    {
        candles_to_trace("XBTUSD", CandleUnit.Month, 1),
        candles_to_trace("XBTUSD", CandleUnit.Day, 1)
    });

chart.WithLayout(new Layout.Layout()
{
    title = String.Format("BTC")
});

chart
```
